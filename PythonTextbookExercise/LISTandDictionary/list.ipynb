```python
empty_list=[]
numbers=[1,2,3,4,5]
fruits=['apple','banana','orange','grape']
mixed=[1, "Hello",3.14, True, [1,2,3]]

print(f"empty list: {empty_list}")
print(f"numbers list: {numbers}")
print(f"fruits list: {fruits}")
print(f"mixed list: {mixed}")

# generating list using list() function
range_list=list(range(1,6))
# In Python, range(1, 6) creates a range object that represents a sequence of numbers starting from 1 up to but not including 6.
string_list=list("Python")

print(f"generated by using range: {range_list}")
print(f"generated by using string: {string_list}")

print(f"length of the fruits list: {len(fruits)}")
```

```python
# list indexing and slicing
numbers=[10,20,30,40,50,60,70,80,90]

print(f"the first element: {numbers[0]}")
print(f"the last element: {numbers[-1]}")
print(f"the second to the last element: {numbers[-2]}")

print(f"the first three elements: {numbers[:3]}")
print(f"the last three elements: {numbers[-3:]}")
print(f"from the third to the fifth elements: {numbers[2:6]}")
print(f"first third fifth elemtns: {numbers[::2]}")
print(f"in reversed order: {numbers[::-1]}")

# # ğŸ” What [::2] Means

# This is a slice with three parts:
#     start: where to begin (default is 0)

#     stop: where to end (default is the end of the list)

#     step: how many items to skip â€” in this case, 2

# So [::2] means:

#     Start from the beginning

#     Go all the way to the end

#     Take every second element
numbers[1:4] = [25, 35, 45]
print(f"ì¼ë¶€ ë³€ê²½ í›„: {numbers}")

fruits = ["ì‚¬ê³¼", "ë°”ë‚˜ë‚˜", "ì˜¤ë Œì§€"]
fruits[1] = "ë”¸ê¸°"
print(f"ìˆ˜ì • í›„: {fruits}")  # ['ì‚¬ê³¼', 'ë”¸ê¸°', 'ì˜¤ë Œì§€']
```
